```
import pandas as pd
import numpy as np

files = [
    "userver_pull_requests_2025-03-02 10-58-15.json",
    "catboost_pull_requests_2025-03-02 10-50-31.json",
    "ydb_pull_requests_2025-03-01 21-26-04.json"
]

source_map = {
    "userver_pull_requests_2025-03-02 10-58-15.json": "userver",
    "catboost_pull_requests_2025-03-02 10-50-31.json": "catboost",
    "ydb_pull_requests_2025-03-01 21-26-04.json": "ydb"
}

df_pull = []

for file in files:
    df = pd.read_json(file)
    df['source'] = source_map.get(file, file)
    df_pull.append(df)

pull_df = pd.concat(df_pull, ignore_index=True)
pull_df.head()

```

```
pull_df.rename(columns={'user_login': 'login'}, inplace=True)
```

```
pull_df.head()
```

```
files_contr = [
    "userver_contributors_2025-03-02 10-58-15.json",
    "catboost_contributors_2025-03-02 10-50-31.json",
    "ydb_contributors_2025-03-01 21-26-04.json"
]

df_contr = {}

for file in files_contr:
    df_contr[file] = pd.read_json(file)

contr_df = pd.concat(df_contr.values(), ignore_index=True)

contr_df.head()
```

```
contr_df.columns
```

```
contr_df.drop(['avatar_url', 'gravatar_id', 'url', 'html_url', 'followers_url', 'following_url', 'gists_url',
       'starred_url', 'subscriptions_url', 'organizations_url', 'repos_url',
       'events_url', 'received_events_url', 'twitter_username', 'blog', 'node_id'], axis=1, inplace=True)
```

```
contr_df.head()
```

```
f_df = pd.merge(pull_df, contr_df, on='login', how='left')
```
